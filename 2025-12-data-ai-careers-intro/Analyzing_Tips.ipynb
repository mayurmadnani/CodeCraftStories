````python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
````

````python
tips.head()
````

````python
tips.info()
````

````python
plt.figure(figsize=(10, 5))
sns.histplot(data=tips, x='tip', bins=40, kde=True, edgecolor='black')
plt.title('Tip Amounts Histogram: Rounding Behavior')
plt.xlabel('Tip Amount ($)')
plt.ylabel('Frequency')
plt.xticks(range(0, int(tips['tip'].max()) + 2))
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
````



```
Tip Amounts Histogram: Rounding Behavior
```

This histogram visualizes the distribution of tip amounts. A histogram is chosen to show the frequency distribution of a single numerical variable. We can observe the most common tip amounts and identify potential rounding behavior, where tips often cluster around whole dollar values or common increments (e.g., $2, $3, $4).

**Conclusion**: The histogram reveals that tip amounts are not uniformly distributed but show peaks at certain values, suggesting a tendency for customers to round their tips, particularly at integer dollar amounts.

````python
plt.figure(figsize=(10, 5))
sns.violinplot(data=tips, x='sex', y='tip', inner='quartile', palette='pastel')
# sns.boxplot(data=tips, x='sex', y='tip',  palette='pastel')
plt.title('Gender vs. Tip Amount: Distribution & Variance')
plt.xlabel('Gender')
plt.ylabel('Tip Amount ($)')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
````



```
Gender vs. Tip Amount: Distribution & Variance
```

This violin plot illustrates the distribution of tip amounts for different genders. A violin plot is ideal for comparing the distribution of a numerical variable across different categories, as it shows both the density estimation (the "violin" shape) and summary statistics (like the inner quartile box plot). This allows for a richer comparison than a simple box plot, revealing the spread and likelihood of different tip amounts within each gender group.

**Conclusion**: The violin plot helps us understand if there are noticeable differences in the typical tip amount and its variability between male and female customers. For instance, we can observe if one gender tends to give higher tips or if their tip amounts are more spread out.

````python
plt.figure(figsize=(10, 5))
sns.boxplot(data=tips, x='smoker', y='tip', palette='coolwarm')
plt.title("Smoker's Premium: Tip Amount Variability")
plt.xlabel('Smoker Status')
plt.ylabel('Tip Amount ($)')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
````


```
Smoker's Premium: Tip Amount Variability
```


This box plot compares the distribution of tip amounts based on whether the customer is a smoker or a non-smoker. A box plot is an excellent choice for comparing the central tendency (median), spread (interquartile range), and potential outliers of a numerical variable across different categorical groups. It clearly shows the median tip, the range of the middle 50% of tips, and any unusually low or high tips.

**Conclusion**: By examining the box plots, we can determine if there's a significant difference in the median tip amount or the overall variability of tips given by smokers versus non-smokers. This can reveal whether smoking status correlates with tipping behavior.
